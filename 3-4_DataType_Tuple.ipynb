## 튜플 다루기

```python
# 인덱싱
t1=(1,2,'a','b')
t1[0]
```

```python
# 슬라이싱
t1[2:4]
```

```python
t1[:2]
```

```python
# 튜플 더하기 : 연결
t1=(1,2,'a','b')
t2=(3,4)

t3=t1+t2
t3
```

```python
# 튜플 곱하기 : 반복
t3=t2*3
t3
```

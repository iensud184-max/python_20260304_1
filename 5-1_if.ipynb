## 3. 다양한 조건을 판단하는 elif

```python
# if/else로만 표현
pocket = ['paper', 'cellphone']
card=True

if 'money' in pocket:
    print('택시를 타고 가라')
else:
    if card:
        print('택시를 타고 가라')
    else:
        print('걸어가라')
```

```python
# elif를 사용
```

```python
if 'money' in pocket:
    print('택시를 타고 가라')
elif card:
    print('택시를 타고 가라')
else:
    print('걸어가라')
```

```python
# if문 한 줄로 작성
```

## 4. 조건부 표현식

```python
# if문
```

```python
score=50

if score >=60:
    message = 'success'
else:
    message = 'fallure'

print(message)
```

```python
# 조건부 표현식conditional expression)
score=90
message='success' if score >=60 else 'fallure'
message
```

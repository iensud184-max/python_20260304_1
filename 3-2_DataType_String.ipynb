## 11. f문자열 포매팅

```python
# 문자열 앞에 f 접두사를 붙이기
```

```python
name='김동환'
age=26
f'나의 이름은 {name}이고, 나이는 {age}살입니다.'
```

```python
# 표현식 사용하기
```

```python
f'나는 내년이면 {age+1}살이 됩니다.'
```

```python
d={'name':'김동환1', 'age':261}
f'나의 이름은 {name}이고, 나이는 {age}살입니다.'
```

```python
d={'name':'김동환1', 'age':261}
f'나의 이름은 {d['name']}이고, 나이는 {d['age']}살입니다.'
```

## 5. 딕셔너리 관련 함수

### keys() : key 리스트 만들기

```python
dic={'name': '김동환', 'phone':'010-2557-3564', 'birth': '1207'}
dic.keys()
```

```python
list(dic.keys())
```

```python
for key in dic.keys():
    print(key)
```

### values() : value 리스트 만들기

```python
dic.values()
```

```python
for value in dic.values():
    print(value)
```

### items() : key, value 쌍 얻기

```python
dic.items()
```

### clear() : key, value 쌍 모두 지우기

```python
dic
```

```python
dic.clear()
dic
```

### get() : key로 value 얻기 

```python
a={'name': '김동환', 'phone': '010-2557-3564', 'birth': '1207'}
```

```python
a['name']
```

```python
a.get('name')
```

```python
print(a.get('age'))
```

```python
a['age']
```

```python
a
```
